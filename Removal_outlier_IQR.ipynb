```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("data/Repair_Motor_vehicle.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
sns.boxplot(x="# ASE Certified Mechanics", data=dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset["# ASE Certified Mechanics"].quantile(0.25)
q3 = dataset["# ASE Certified Mechanics"].quantile(0.75)

q3
q1
```

```python
dataset.describe()
dataset.shape
```

```python
IQR = q3-q1

min_range = q1 - (1.5*IQR)


max_range = q3 + (1.5*IQR)

min_range
max_range
```

```python
new_dataset = dataset[dataset["# ASE Certified Mechanics"]<=max_range]
new_dataset
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="# ASE Certified Mechanics", data=new_dataset)
plt.show()
```
