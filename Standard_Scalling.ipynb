# Feature Scalling 

#### Two types of feature scalling
- Standardization
- Normalization

## Standardization Formula

Xnew = (Xi - Xmean) / Standard Deviation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv("data/Repair_Motor_vehicle.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.displot(dataset["# ASE Certified Mechanics"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

ss.fit(dataset[["# ASE Certified Mechanics"]])

```

```python
dataset["# ASE Certified Mechanics ss"] = pd.DataFrame(ss.transform(dataset[["# ASE Certified Mechanics"]]), columns=["X"]) 
dataset.head()
```

```python
dataset.describe()
```

```python
sns.displot(dataset["# ASE Certified Mechanics"])
plt.show()
```
